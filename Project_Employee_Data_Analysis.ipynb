### 1. How many are there in each Team and the percentage splitting with respect to the total employees.

```python
total_employees = len(df)
df1 = df["Team"].value_counts()
print((df1/total_employees*100).round(2))
```

### 2. Segregate the employees w.r.t different positions.

```python
df2 = df.groupby(['Position', 'Name']).size()
print(df2.to_string())
```

### 3. Find from which age group most of the employees belong to.

```python
# Create the histogram plot to analyse frequency distribution for employee age
plt.hist(df['Age'])

# add features
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram Plot')

# Display the plot
plt.show()
```

### 4. Find out under which team and position, spending in terms of salary is high. 

```python
df3 = df.sort_values(by='Salary', ascending = False)
print("  The Team : '{}' \n  The Position : '{}'".format(df3.Team.iloc[0],df3.Position.iloc[0]))
```

### 5. Find if there is any correlation between age and salary , represent it visually. 

```python
df4 = df[['Age', 'Salary']]
df5 = df4.corr()
df5
```

```python
# Load dataset
df4 = df[['Age', 'Salary']]

# Compute correlation matrix
corr = df4.corr()

# Create heatmap
sns.heatmap(corr, annot=True, cmap="flare", center=0, linewidth=1)

# Add title
plt.title("Correlation between Age and Salary")

# Show plot
plt.show()
```
